```python
import pymongo

cli = pymongo.MongoClient(host='localhost', port=27017)
db = cli['douban']
collection = db['movies']

datas = collection.find()
```

```python
import pandas as pd

datas = pd.DataFrame(list(datas))

# 删除_id
datas.drop('_id', axis=1, inplace=True)
datas.d()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

font = {
    'size': 12,
    'family': 'simhei',
    'weight': 'normal'
}

mpl.rc('font', **font)
```

```python
years_movie_num = datas.groupby('year').count().reset_index().sort_values('year', ascending=False)[3:50]
years_movie_num
fig = plt.figure(figsize=(20,8))
sns.lineplot(x=years_movie_num['year'], y=years_movie_num['name'])
plt.xticks(rotation=45)
plt.title('豆瓣收录电影每年数量')
plt.show()
```
